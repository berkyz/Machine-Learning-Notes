**PANDAS KÜTÜPHANESİ**

Pandas, Python programlama dili için geliştirilmiş, yüksek performanslı ve kullanımı kolay veri yapıları sunan açık kaynak kodlu bir kütüphanedir. Temelleri NumPy kütüphanesi üzerine kurulmuştur ve özellikle veri analizi ile veri ön işleme süreçlerinde vazgeçilmez bir araçtır. Yazılım projelerine genellikle import pandas as pd komutu ile dahil edilir.

**1. Temel Veri Yapıları**

Pandas, verileri işlemek için iki temel yapı sunar: Seriler (Series) ve DataFrame yapıları.

**A. Seriler (Series)**

Seriler, herhangi bir veri tipini (sayısal, metinsel vb.) saklayabilen tek boyutlu dizilerdir.

**İndis Yapısı:** Serideki her eleman bir indis değerine sahiptir. Varsayılan olarak ilk elemanın indisi sıfırdır.

**Esneklik:** Seriler oluşturulurken elemanlara manuel olarak etiket verilebilir veya mevcut sözlük (dictionary) yapıları kolayca seriye dönüştürülebilir.


**Veri Tipi:** Bir seri içinde hem sayısal hem de sayısal olmayan değerler bir arada bulunabilir; bu durumda veri tipi genellikle object olarak tanımlanır.

**B. DataFrame Yapıları**

DataFrame, satır indis etiketleri ve sütun adlarını içeren, Excel tablolarına benzeyen çok boyutlu ve elektronik tablo yapısına sahip bir nesnedir.

**Çok Boyutlu Temsil: ** Serilerin bir araya gelmesiyle oluşur ve her bir sütun farklı bir veri tipine sahip olabilir.

**Veri Okuma Kapasitesi:** Farklı formatlardaki dosyaları (CSV, JSON, HTML, Excel, SQL vb.) okuyarak doğrudan DataFrame yapılarına dönüştürebilir.


**2. Veri Analizi ve Fonksiyonlar**

Pandas, veri setini tanımak ve istatistiksel analizler yapmak için güçlü fonksiyonlar sunar:

**Veri Tanıma:** info() fonksiyonu ile veri setinin içeriği (bellek kullanımı, veri tipleri) hakkında bilgi alınabilir. head() ve tail() fonksiyonları ise verinin başındaki veya sonundaki kayıtları hızlıca incelemeye yarar.


**Filtreleme ve Sorgulama:** Veri seti üzerinde koşullu sorgulamalar yapılarak belirli kriterlere (örneğin; puanı 95'ten büyük binalar) uyan veriler kolayca ayıklanabilir.

**Korelasyon Analizi:** Öznitelikler arasındaki istatistiksel ilişkiyi belirlemek için corr() fonksiyonu kullanılır. İlişki katsayısı 1'e yaklaştıkça ilişkinin kuvveti artar.

**3. Matematiksel İşlemler**

Pandas yapıları üzerinde NumPy fonksiyonları doğrudan uygulanabilir. Veri setindeki elemanların mutlak değeri (Abs), karekökü (Sqrt) veya kareleri (Square) gibi işlemler her bir eleman için otomatik olarak hesaplanabilir. Ayrıca sayıları üste (Ceil) veya alta (Floor) yuvarlama gibi hassas işlemler veri manipülasyonu sürecinde aktif olarak kullanılır

2. Sözlükten (Dictionary) Seri Oluşturma ve Filtreleme
Python sözlük yapısını seri nesnesine dönüştürme ve veri üzerinde sorgulama işlemleri.

```python
# Sözlükten seri oluşturma
notlar = {"Matematik": 70, "Bilgisayar": 90, "Elektronik": 96, "İstatistik": 55}
seri_notlar = pd.Series(notlar)
print("Notlar Serisi:\n", seri_notlar)
```

```python
# Seri içinde anahtar kontrolü
print("\nFizik notu var mı?:", 'Fizik' in seri_notlar)
```

```python
# Matematiksel işlemler ve Fonksiyon kullanımı
print("\nNotların %5 fazlası:\n", seri_notlar * 1.05)
print("\nNotların Karekökü:\n", np.sqrt(seri_notlar))
```

```python
# Koşullu Filtreleme
sayi_serisi = pd.Series([10, 15, 20, 25, 30, 35])
print("\n22'den büyük elemanlar:\n", sayi_serisi[sayi_serisi > 22])
```

```python
# Benzersiz değerler ve Frekans sayımı
karisik_seri = pd.Series([10, 15, 20, 25, 20, 10, 15, 45, 50, 20])
print("\nBenzersiz Değerler:", karisik_seri.unique())
print("Değerlerin Geçme Sayısı:\n", karisik_seri.value_counts())
```

5. Korelasyon Analizi ve Isı Haritası (Heatmap)
Öznitelikler arasındaki istatistiksel ilişkiyi belirleme ve görselleştirme.

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Korelasyon matrisi hesaplama (Sayısal sütunlar için)
# korelasyon = veriseti.corr()
# print(korelasyon)

# Seaborn ile Isı Haritası (Heatmap) oluşturma
# sb.heatmap(korelasyon, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.01)
# plt.show()
```

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

# 1. Veri setini yükleyelim
veriseti = pd.read_excel('EnergyStarScore.xlsx', 'Sayfa1')

# 2. Korelasyon matrisi hesaplama
korelasyon = veriseti.corr(numeric_only=True)

# --- YENİ EKLENEN DETAYLI PRİNT BÖLÜMLERİ ---

print("=== VERİ SETİ KORELASYON ÖZETİ ===\n")

# A. Boyut ve Sütun Bilgisi
# matrisin .shape[0] özelliği kaç satır/sütun olduğunu verir.
print(f"1. Analiz Edilen Sayısal Değişken Sayısı: {korelasyon.shape[0]}")
print(f"2. İşleme Alınan Sütunlar: {list(korelasyon.columns)}\n")

print("3. Genel Korelasyon Matrisi:")
print(korelasyon)
print("\n")

# B. En Güçlü İlişkileri Filtreleme
# unstack() ile tabloyu satır satır çözüp tek bir liste haline getiriyoruz.
# Böylece çiftleri daha kolay sıralayabiliriz.
korelasyon_ciftleri = korelasyon.unstack()

# Bir değişkenin kendisiyle korelasyonu her zaman 1.0'dır. Bu gereksiz bilgiyi filtreleyip, 
# drop_duplicates() ile aynı çiftin tekrar edilmesini ("A-B" ile "B-A") engelliyoruz.
filtrelenmis_ciftler = korelasyon_ciftleri[korelasyon_ciftleri < 1.0].drop_duplicates()

# Değerleri büyükten küçüğe doğru sıralıyoruz
sirali_ciftler = filtrelenmis_ciftler.sort_values(ascending=False)

print("4. EN GÜÇLÜ POZİTİF İLİŞKİLER (Doğru Orantı):")
print("Biri artarken diğeri de artma eğiliminde olan ilk 3 çift:")
# .head(3) en üstteki (en yüksek pozitif) 3 değeri verir
print(sirali_ciftler.head(3))
print("\n")

print("5. EN GÜÇLÜ NEGATİF İLİŞKİLER (Ters Orantı):")
print("Biri artarken diğeri azalma eğiliminde olan ilk 3 çift:")
# .tail(3) en alttaki (en güçlü negatif) 3 değeri verir
print(sirali_ciftler.tail(3))
print("==================================\n")


# 3. Seaborn ile Isı Haritası (Heatmap) oluşturma
plt.figure(figsize=(10, 8)) 
sb.heatmap(korelasyon, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.01)

plt.title("Öznitelikler Arasındaki Korelasyon Isı Haritası")
plt.show()
```
