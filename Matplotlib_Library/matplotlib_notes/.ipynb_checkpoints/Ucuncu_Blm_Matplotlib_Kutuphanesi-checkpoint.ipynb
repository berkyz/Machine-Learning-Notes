**Matplotlib Kütüphanesi Nedir?**

Matplotlib, Python programlama dili için geliştirilmiş, iki veya üç boyutlu grafikler oluşturmaya yarayan güçlü bir görselleştirme kütüphanesidir. Makine öğrenmesi süreçlerinde veri analizi yaparken, verilerin grafikler aracılığıyla dağılımını görmek, hangi algoritmanın (örneğin doğrusal mı yoksa polinom regresyon mu) kullanılacağına karar vermek açısından kritik bir öneme sahiptir.

**Temel Özellikleri ve Avantajları**

**Çıktı Formatları:** Hazırlanan grafikler PNG, PDF ve SVG gibi popüler resim formatlarında kolayca kaydedilebilir.

**LaTeX Desteği:** Grafiklerin üzerine matematiksel formülleri ve metinleri eklemek için LaTeX formatı kullanılabilir.

**Esneklik:** pyplot modülü sayesinde grafik üzerindeki tüm detaylar (eksenler, başlıklar, renkler) programlanabilir.

**Grafiği Şekillendiren Temel Fonksiyonlar**

Bir grafik oluştururken kullanılan başlıca fonksiyonlar ve görevleri şunlardır:

**plt.axis():** Eksenlerin değer aralığını belirler. Örneğin plt.axis([2009, 2015, 30, 80]) komutu x eksenini 2009-2015, y eksenini ise 30-80 aralığına sabitler. Otomatik ayar için plt.axis('auto') kullanılabilir.


**plt.title():** Grafiğe genel bir başlık ekler.

**plt.xlabel() ve plt.ylabel():** Sırasıyla x ve y eksenlerinin isimlerini (etiketlerini) tanımlar.

**plt.plot():** Verileri grafiğe döker. İlk iki parametre x ve y verilerini içeren listelerdir; üçüncü parametre ise işaretçinin şeklini ve rengini belirler (Örn: 'ro' kırmızı yuvarlak, 'g-' yeşil tek çizgi).


**plt.show():** Hazırlanan grafiğin ekranda görüntülenmesini sağlar.

**Not:** Verilerin noktalar halinde gösterildiği grafik türü genel olarak saçılım grafiği (scatter plot) olarak adlandırılır.

1. Temel Saçılım Grafiği
Bu grafik, Türkiye'nin taşıma sektörüne ait sera gazı emisyon değerlerini gösterir.

```python
import matplotlib.pyplot as plt

# Eksen aralıklarını belirleme: [xmin, xmax, ymin, ymax]
plt.axis([2009, 2015, 30, 80])
plt.title("Türkiye'nin Taşıma Sektörüne Ait Sera Gazı Emisyon Değerleri")
plt.xlabel("Yıllar")
plt.ylabel("Emisyon Değeri(bin Gg)")

# 'm*' parametresi eflatun rengi ve yıldız işaretini temsil eder
plt.plot([2010, 2011, 2012, 2013, 2014], [45, 47, 62, 68, 73], 'm*')
plt.show()
```

2. Metin ve Izgara Eklenmiş Grafik
Grafiğin belirli koordinatlarına text() fonksiyonu ile açıklama eklenmiş halidir.

```python
import matplotlib.pyplot as plt

plt.axis([2009, 2015, 30, 80])
plt.title("Türkiye'nin Taşıma Sektörüne Ait Sera Gazı Emisyon Değerleri")

# Yazı tipi ve renk özelleştirmeleri
plt.xlabel("Yıllar", fontsize=15, fontname='DejaVu Sans', color='red')
plt.ylabel("Emisyon Değeri (bin Gg)", fontsize=15, fontname='DejaVu Sans', color='red')

# Belirli koordinatlara metin ekleme
plt.text(2009.7, 47, "En Düşük", color='Green', style='italic', weight='bold', size='7')
plt.text(2013.7, 75, "En Yüksek", color='Blue')

plt.grid() # Izgara ekleme
plt.plot([2010, 2011, 2012, 2013, 2014], [45, 47, 62, 68, 73], 'm*')
plt.show()
```

3. Çok Serili Zaman Grafiği
İki farklı makineye ait sıcaklık değişimlerini zaman bazlı gösterir

```python
import matplotlib.pyplot as plt
import datetime

plt.axis('auto')
plt.title("Makina Sıcaklık Takip Grafiği", fontsize=14)
plt.xlabel("Saat", fontsize=12)
plt.ylabel("Sıcaklık (C)", fontsize=12)

y1 = [120, 122, 124.5, 126, 126.4]
y2 = [100, 102, 103.5, 104, 105]

# Zaman verisi üretme (her veri için 1 dakika ekler)
x = [datetime.datetime.now() + datetime.timedelta(minutes=i) for i in range(len(y1))]

plt.grid()
plt.plot(x, y1, 'sr', label='Makina 1') # Kırmızı kare
plt.plot(x, y2, 'm*', label='Makina 2') # Eflatun yıldız

# loc=2 sol üst köşeyi temsil eder
plt.legend(loc=2)
plt.gcf().autofmt_xdate() # Tarih formatını otomatik düzenle
plt.show()
```

4. Eşik Değer Gösterimi
Belirli bir değerin (9) altındaki ve üstündeki verilerin farklı renklerle gösterilmesi

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2, 3, 4, 5, 7, 9, 12, 13, 14, 10, 7, 11, 14, 15])
x = np.arange(len(y))
esik = 9

plt.title("Eşik Değer Aşımı", fontsize=14)
plt.plot(x, y, color='black')

# Maskeleme ile renk ayrımı
altEsik = y < esik
plt.scatter(x[altEsik], y[altEsik], color='black')
ustEsik = np.logical_not(altEsik)
plt.scatter(x[ustEsik], y[ustEsik], color='red')

# Yatay eşik çizgisi çizme
plt.axhline(esik, color='red', linestyle='--')
plt.show()
```

5. Matematiksel Fonksiyon Çizimi
LaTeX formülleri kullanılarak bir polinomun grafiğe aktarılması.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 1000)
y = (x**3) + 4*x + 6

# bbox ile formülün etrafına kutu eklenmiştir
plt.text(-2, 100, r'$y=x^3+4*x+6$', fontsize=12,
         bbox={'facecolor': 'blue', 'alpha': 0.2})

plt.grid()
plt.plot(x, y)
plt.show()
```

6. Histogram Grafiği
80 öğrencinin final notlarının istatistiksel dağılımı

```python
import matplotlib.pyplot as plt
import numpy as np

X = [56,52,48,16,45,20,52,52,36,60,70,52,36,56,20,80,56,36,44,36,
     56,48,68,56,80,64,60,12,56,60,36,72,24,60,60,64,60,60,60,64,
     52,76,52,68,40,72,56,46,32,34,64,60,40,60,60,60,28,56,72,52,
     64,28,64,52,52,56,44,52,60,64,40,36,48,48,56,46,56,76,48,44]

plt.title("Bilgisayar Programlama Dersi Final Sınavı Histogramı")
plt.xlabel('Notlar')
plt.ylabel('Öğrenci Sayısı')
plt.axis([0, 100, 0, 15])
plt.grid(True)

# bins=15 ile veri 15 gruba ayrılır
plt.hist(X, bins=15, facecolor='blue', alpha=0.75)
plt.show()
```

7. 3D Yüzey Grafiği
Karmaşık bir fonksiyonun üç boyutlu görselleştirilmesi

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

def fonksiyon(x, y):
    return ((x**2 + y**2) / 2) * np.exp(-(x**2 - y**3) / 4)

x = np.arange(-1.0, 1.0, 0.01)
y = np.arange(-1.0, 1.0, 0.01)
a, b = np.meshgrid(x, y)
z = fonksiyon(a, b)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

yuzey = ax.plot_surface(a, b, z, rstride=1, cstride=1, cmap=plt.cm.RdBu,
                       linewidth=0, antialiased=False)

fig.colorbar(yuzey, shrink=0.6)
plt.title(r'$z=((x^2+y^3)/2) e^{-(x^2-y^2)/4}$')
plt.show()
```

8. Pasta Grafiği
Enerji sektörü salınımlarının yüzdesel dağılımı

```python
import matplotlib.pyplot as plt

etiketler = ['Türkiye', 'Almanya', 'Yunanistan', 'Fransa', 'Danimarka', 'İspanya']
veriler = [339104, 762338, 75206, 319636, 36738, 238091]
renkler = ['#ff9999','#66b3ff', '#99ff99','#ffcc99', '#bbadf9', '#a9ebed']
ayrikParca = (0.1, 0, 0, 0, 0, 0) # Türkiye dilimini ayırır

plt.pie(veriler, labels=etiketler, colors=renkler, explode=ayrikParca, autopct='%1.1f%%')
plt.axis('equal') # Grafiği tam daire yapar
plt.title("Avrupa'da 2014 Yılı Enerji Sektörüne Ait Seragazı Salınımları")
plt.show()
```
